```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.metrics import f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import ComplementNB
```

```python
df = pd.read_csv('cleared_train_v4.csv',parse_dates=['datetime'])
```

```python
crash_stat = df.groupby(['road_id', 'road_km']).agg({'bin_target': 'sum'})\
  .sort_values(by='bin_target', ascending=False)\
  .reset_index()
```

```python
#статистика по авариям
crash_stat.head()
```

```python
#берем эти участки для предсказания, остальные участки зануляем
condition1 = (crash_stat.road_id == 9) & (crash_stat['bin_target'] > 400)
condition2 = (crash_stat.road_id == 14) & (crash_stat['bin_target'] > 100)

```

```python
stat=crash_stat[condition1 | condition2].sort_values(by=['road_id', 'bin_target'], ascending=False).reset_index(drop=True)
```

```python
len(stat)
```

```python
type(stat)
```

```python
sets=[]
for i in range(len(stat)):
    sets.append({stat['road_id'][i], stat['road_km'][i]})
```

```python
#считываем предсказания нашей модели
submit=pd.read_csv('submit_4.csv',parse_dates=['datetime'])
```

```python
s=[]
for i in range(len(submit)):
    if {submit['road_id'][i], submit['road_km'][i]} in sets:
        s.append(submit.iloc[i])
```

```python
#экспортируем отредактированный сабмит
pd.DataFrame(s).to_csv('s1_16.csv', index=None)
```
